```python
from gql import gql, Client
from gql.transport.requests import RequestsHTTPTransport
sample_transport=RequestsHTTPTransport(
   url='https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2',
   verify=True,
   retries=5,
)
client1 = Client(
   transport=sample_transport
)
query = gql('''
query trades{
  swaps(where:{pair: "0xd3d2e2692501a5c9ca623199d38826e513033a17"}, first:100, block: {number:11009390}) {
   timestamp
    pair {
      id
      volumeToken0
      volumeToken1
      volumeUSD
      token0 {
        
        symbol
        name
        
      }
      token1 {
        
        symbol
        name
        
      }
  
  
      
    }
   
    
  }
}
''')
new1 = client1.execute(query)
print(new1)
```
